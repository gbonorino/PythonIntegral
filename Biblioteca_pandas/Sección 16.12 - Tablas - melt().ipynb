#### Conversin

```python
# Creamos una tabla en formato ancho
import pandas as pd

datos_ancho= {'Nombre':['Mara','Luis','Carmen'],
        'Matemticas':[8.5, 7, 3.5], 'Economa':[8, 6.5, 5],'Programacin':[6.5, 4, 9]}
df=pd.DataFrame(datos_ancho)
print(df)
```

```python
# La convertimos a formato largo con melt()
#df_largo = df.melt()
#df_largo = df.melt(id_vars='Nombre') 
df_largo = df.melt(id_vars='Nombre', var_name='Asignatura',value_name='Nota')
print(df_largo)
```
