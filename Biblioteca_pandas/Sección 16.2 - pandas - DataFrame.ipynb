#### Crear DataFrames

```python
# Con listas, una columna
import pandas as pd

lista = [4, 5, 6, [7, 8, 9], ['uno', 'dos']]
df = pd.DataFrame(lista, columns = ["Columna"])   
print(type(df))
print(df)
```

```python
# Con listas, tres columnas
import pandas as pd 

lista = [['Tomas', 'Reacher', 25], ['Cris', 'Salvio', 30], 
       ['Nick', 'Wilson', 26], ['Julio', 'Sach', 22]] 
df = pd.DataFrame(lista, columns =['Nombre', 'Apellido', 'Edad']) 
df
```

```python
# A partir de Series
import pandas as pd

d =  pd.DataFrame({ 'Nombre': pd.Series(['Alicia','Roberto','Carlos']), 
                  'Edad': pd.Series([21,25,23]) } )
print(d)
```

```python
# A partir de un archivo CSV
import pandas as pd

data = pd.read_csv('Cordoba_precip.csv') 
print(data)
type(data)
```
