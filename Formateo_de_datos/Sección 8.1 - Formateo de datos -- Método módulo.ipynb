#### Operador módulo

```python
print("Cantidad: %d, Precio unitario: %7.3f, Total: %.2f" % (500, 2.578, 500*2.578))
```

```python
# Insertar un único valor
print("Este es un curso de %s " % "Python")
```

```python
# Múltiples valores en una tupla
print("%d %s cuestan $%.2f" % (6, "bananas", 159.00))
```

```python
# Insertar símbolo de porcentaje
print('$%.2f a interés del %.f%% en %d días rinden $%.2f' % (100, 10, 30, 25.50))
```

#### Formateo de números

```python
# Variar el número de decimales
import math

print("pi = %.0f " % math.pi)
```

```python
# Variar el número de espacios
print(" pi = %17.5f " % 3.14159)
```
