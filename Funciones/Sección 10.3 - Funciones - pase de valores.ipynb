#### Pase de valores

Mecanismos de pase de valor <br>
	Por posicin <br>
	Por palabras clave <br>
	Por valor predefinido <br>
	Nmero indefinido

#### Por posicin

```python
def suma(a, b):
    return a*2 + b*8

suma(4, 6)
```

#### Por palabra clave

```python
def pitagoras(lado1, lado2):
    return (lado1**2 + lado2**2)**0.5

pitagoras(lado2=6,lado1=4)
```

#### Prefijando el valor de un parmetro

```python
print("Usa el valor prefijado")
def suma(a, b = 100):
    return a*2 + b*8

print(suma(4))
```

```python
print("No usa el valor prefijado")
def suma(a, b = 100):
    return a*2 + b*8

print(suma(4, 6))
```

#### Nmero indefinido
La sintaxis de print es:<br>
print(*objects, ms otros argumentos)

```python
print()
print("Un parametro")
print("Parametro_1, Parametro_2, Parametro_3")
```

```python
def suma(*args):
    resultado = 0
    for i in args:
        resultado += i
    return resultado

print("Dos argumentos")
print(suma(4, 6))
print()
print("Tres argumentos")
print(suma(4, 6, 8))
```

```python
# Con palabras clave
def agenda(**kwargs):
    for clave, valor in kwargs.items():
        print(clave, valor)

print(agenda(Nombre = "Augusto", Apellido = "Bonorino", Edad = 21))
```
